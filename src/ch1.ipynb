### 1.1.3 NumPyとMatplotlibの基本的な使い方

- 状態ベースインターフェース (state-based interface)

- List 1.1: matplotlibExample_1.py

$$
x(t)=e^t \sin 10t
$$

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-2, 2, 100)
plt.plot(t, np.exp(t)*np.sin(10*t))
# 装飾
plt.title(r'sample graph $x(t)=e^t \sin 10t$')
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

- オブジェクト指向インターフェース

- List 1.2: matplotlibExample_2.py

$$
\begin{cases}
    x_1(t)=\sin t \\
    x_2(t)=\cos t + \frac{t}{5}\\
    x_3(t)=\sin^2 t \\
    x_4(t)=\sqrt{|t|} \cos t
\end{cases}
$$

```python
t = np.linspace(-5, 5, 1000)
x=[
    np.sin(t),
    np.cos(t)+t/5,
    np.sin(t)**2,
    np.sqrt(np.abs(t))*np.cos(t),
]

fig = plt.figure()

for i, f in enumerate(x, start=1):
    ax = fig.add_subplot(2,2,i)
    ax.plot(t, f)
    ax.set_ylim(-2.5, 2.5)

plt.show()
```
